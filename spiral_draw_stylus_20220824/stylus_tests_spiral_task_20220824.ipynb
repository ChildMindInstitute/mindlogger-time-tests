```python
import pandas as pd
import os

report_table = "/Users/arno/Software/mindlogger-time-tests/spiral_draw_stylus_20220824/report.csv"
dir_name = "/Users/arno/Software/mindlogger-time-tests/spiral_draw_stylus_20220824/drawing-responses-Thu Aug 25 2022"
#"line_number","x","y","UTC_Timestamp","seconds","epoch_time_in_seconds_start"

report_df = pd.read_csv(report_table)
nduplicate_times = []
nduplicate_coordinates = []
names = report_df['name']
filenames = report_df['filename']
for filename in filenames:
    file_df = pd.read_csv(os.path.join(dir_name, filename + '.csv'))
    time_df = file_df[file_df.duplicated('UTC_Timestamp', keep=False)].sort_values('UTC_Timestamp')
    xy_df = file_df[file_df.duplicated(['x', 'y'], keep=False)].sort_values('UTC_Timestamp')

    nduplicate_times.append(len(time_df))
    nduplicate_coordinates.append(len(xy_df))

    print('File: {0}'.format(filename))
    print('{0} duplicate time stamps'.format(len(time_df)))
    print(time_df[['line_number', 'seconds', 'x', 'y']])
    print('{0} duplicate coordinates'.format(len(xy_df)))
    print(xy_df[['line_number', 'seconds', 'x', 'y']])

data_duplicates = {'Duplicate time stamps':nduplicate_times,
                   'Duplicate coordinates':nduplicate_coordinates,
                   'Names':names}
duplicates_df = pd.DataFrame(data_duplicates)
```

```python
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print('Duplicate time stamps')
    print(duplicates_df['Duplicate time stamps'].to_string(index=False))
    print('Duplicate coordinates')
    print(duplicates_df['Duplicate coordinates'].to_string(index=False))
    print('Duplicates')
    print(duplicates_df.to_string(index=False))
```
