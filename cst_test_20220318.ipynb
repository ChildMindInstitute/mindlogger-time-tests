# Critical Stability Task Plots

## 1a. Do nothing: calibration

```python
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
columns = ['lambda_value','lambda_slope','stim_position','timestamp','user_position']
```

```python
base_dir = '/Users/arno/Software/mindlogger-time-tests/'
csv_file = base_dir + 'input/cst_test_20220318/do-nothing/6234941ade7a36762057f282_calibration-phase.csv'
#csv_file = base_dir + 'input/cst_test_20220318/tap-once/623496a8de7a36762057f283_calibration-phase.csv'
#csv_file = base_dir + 'input/cst_test_20220318/try-center/6234997ade7a36762057f284_calibration-phase.csv'
df = pd.read_csv(csv_file, usecols=columns)
print("Contents in csv file:\n", df)
plt.plot(df.timestamp, df.stim_position, label='stimulus position')
plt.plot(df.timestamp, df.user_position, label='user position')
plt.title('Do nothing: calibration')
plt.xlabel ('Time (ms)')
plt.ylabel ('Position')
plt.legend()
plt.grid()
plt.show()
```

```python
df = pd.read_csv(csv_file, usecols=columns)
plt.plot(df.timestamp, df.lambda_value)
plt.title('Do nothing: calibration -- Lambda value over time')
plt.xlabel ('Time (ms)')
plt.ylabel ('Lambda value')
plt.grid()
plt.show()

df = pd.read_csv(csv_file, usecols=columns)
plt.plot(df.timestamp, df.lambda_slope)
plt.title('Do nothing: calibration -- Lambda slope over time')
plt.xlabel ('Time (ms)')
plt.ylabel ('Lambda slope')
plt.grid()
plt.show()
```
